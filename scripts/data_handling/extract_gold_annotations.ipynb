# Extract Gold Phoneme Annotations
Loads `data/eval/annotations_cache.json` and displays each entry's gold phoneme sequence as a space-separated string labeled by audio ID.

```python
import json
from pathlib import Path

PROJECT_ROOT = Path().resolve()
while not (PROJECT_ROOT / ".git").exists():
    PROJECT_ROOT = PROJECT_ROOT.parent

CACHE_PATH = PROJECT_ROOT / "data" / "eval" / "annotations_cache.json"
```

```python
with open(CACHE_PATH) as f:
    cache = json.load(f)

for entry in cache:
    audio_id = entry["audio_id"]
    gold_str = " ".join(entry["gold"])
    print(f"{audio_id}: {gold_str}")
```
