# Goal
Retrained English ASR outputs somewhat different strings than the original. these were used to synthesize data and those differing should be resynthesized.

# Setup

```python
# Imports
from datasets import load_from_disk
```

```python
from pathlib import Path

# Walk up from the notebook's location to find the project root
PROJECT_ROOT = Path().resolve()
while not (PROJECT_ROOT / ".git").exists(): # anchor project root to .git
    PROJECT_ROOT = PROJECT_ROOT.parent

DATA_PATH = PROJECT_ROOT / "data" / "synthetic" / "paired_old" 
UPDATED_PATH = PROJECT_ROOT / "data" / "synthetic" / "paired"
```

# Load Data

```python
old_ds = load_from_disk(DATA_PATH)
new_ds = load_from_disk(UPDATED_PATH)
```

```python
import pandas as pd

old_asr = old_ds["English ASR transcriptions"]
new_asr = new_ds["English ASR transcriptions"]

new_asr_collapsed = [word.replace(" ", "") for word in new_asr]

diffs = [
    {"index": i, "old": old_asr[i], "new": new_asr_collapsed[i]}
    for i in range(len(old_asr))
    if old_asr[i] != new_asr_collapsed[i]
]

print(f"Total differing rows: {len(diffs)} / {len(old_asr)}")
pd.DataFrame(diffs)
```
