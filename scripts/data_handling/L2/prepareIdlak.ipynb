# Prepare idlak data for use in a dataloader

```python
import resample16k, os
import pandas as pd
from bs4 import BeautifulSoup
#from pathlib import Path
```

## create structured csv
create csv with two columns: audio url and transcription (phonetic) to be used by dataloader

```python
with open("../../data/idlak/transcription/text.xml", "r", encoding="utf-8") as idlak_file:
    contents = idlak_file.read()
```

```python
soup = BeautifulSoup(contents, "xml")
```

```python
print(soup.prettify())
```

```python
data = {}
```

```python
for file in soup.find_all("fileid"):
    name = file["id"]
    transcript = file.string
    data["data/idlak/audio/"+name+".wav"] = transcript
```

```python
data
```

```python
df = pd.DataFrame(list(data.items()))
```

```python
df.head()
```

```python
df.columns = ["audio", "transcription"]
```

```python
df.head()
```

```python
df.to_csv("../../data/idlak/transcription/dataset.csv",index=False)
```
