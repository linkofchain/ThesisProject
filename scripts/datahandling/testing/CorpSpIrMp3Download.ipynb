# Get audio from webpage

## get webpages

```python
# https://www.w3schools.com/python/module_requests.asp
# imports
import re, requests
from bs4 import BeautifulSoup
```

```python
mainPage = "https://www.scss.tcd.ie/elaine.ui.dhonnchadha/comhra/index.utf8.html"
dataFolder = "data/onlineCorpusOfSpokenIrish/speech/"
page_request = requests.get(mainPage)
```

```python
page_content = page_request.text
```

```python
page_content
```

```python
soup = BeautifulSoup(page_content, 'html.parser')
print(soup.prettify())
```

## get mp3s from start page

```python
# examples of file names
# https://www.scss.tcd.ie/elaine.ui.dhonnchadha/comhra/index.utf8.html
# https://www.scss.tcd.ie/elaine.ui.dhonnchadha/comhra/irgl0001.utf8.html
# https://www.scss.tcd.ie/elaine.ui.dhonnchadha/comhra/media/irgl0001.mp3
pagePattern = "(\w{4}\d{4})\.utf8\.html"
```

```python
# gets mp3 from page url.
def get_mp3_url(url, page):
    return re.sub("\/index\.utf8\.html", "/media/"+page+".mp3", url)

pages = re.findall(pagePattern,page_content)
```

```python
#save mp3s to data folder
for page in pages:
    mp3URL = get_mp3_url(mainPage,page)
    mp3Data = requests.get(mp3URL).content
    with open(dataFolder+page+".mp3","wb") as file: #need b for byte writing
        file.write(mp3Data)
```
